```python
import random
import time
from kafka import KafkaProducer
import json

def generate_random_location():
    latitude = random.uniform(-90, 90)
    longitude = random.uniform(-180, 180)
    return {"latitude": latitude, "longitude": longitude}

def send_truck_location(producer, topic, partition, truck_id):
    while True:
        location = generate_random_location()
        location_data = {"truck_id": truck_id, "location": location}
        producer.send(topic, value=location_data, partition=partition)
        print(f"Truck {truck_id} sent location: {location}")
        time.sleep(1)

def main():
    num_trucks = 1
    topic = 'TruckLocations'

    producer = KafkaProducer(
        bootstrap_servers=['localhost:29092'],
        value_serializer=lambda v: json.dumps(v).encode('utf-8')
    )

    for truck_id in range(num_trucks):
        partition = truck_id  # Assign each truck its own partition
        send_truck_location(producer, topic, partition, truck_id)

if __name__ == "__main__":
    main()
```
